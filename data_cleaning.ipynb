### Checking for null values

```python
df.isna().sum()
```

```python
def calculate_null_percentages(df):
    # Calculate the number of null values for each column
    null_counts = df.isnull().sum()
    
    # Calculate the percentage of null values for each column
    null_percentages = (null_counts / len(df)) * 100
    
    return null_percentages

# Example usage with a DataFrame 'df'
# Assuming 'df' is your DataFrame with 4109 entries
null_percentages = calculate_null_percentages(df)

# Print the null percentages
print(null_percentages)
# Set the threshold percentage
threshold = 30.0

# Identify columns to drop
columns_to_drop = null_percentages[null_percentages > threshold].index.tolist()

print("Columns to drop:", columns_to_drop)

# Assume df is your DataFrame
# Drop the columns from the DataFrame
df = df.drop(columns=columns_to_drop)
```

```python
from data_cleaner import DataCleaner  # Import the data_cleaner.py

# Create an instance of DataCleaner with your existing DataFrame
cleaner = DataCleaner(df)

# Set the threshold percentage
threshold = 15.0

# Drop columns based on the null value threshold
columns_dropped_by_threshold = cleaner.drop_columns_by_null_threshold(threshold)
print(f"Columns dropped by threshold: {columns_dropped_by_threshold}")

# Now, you can use cleaned_df for further analysis (if needed)
cleaned_df = cleaner.get_clean_data()
cleaned_df.head()  # Display the first few rows of the cleaned dataframe
```

### Changing datatypes

>Changing the datatype of RESDATE column to datetime format

```python
print(cleaned_df['RESDATE'].value_counts())
print('\n')
print(cleaned_df['FAILDATE'].value_counts())
```

```python
cleaned_df['RESDATE'] = pd.to_datetime(cleaned_df['RESDATE'], errors='coerce')
cleaned_df['FAILDATE'] = pd.to_datetime(cleaned_df['FAILDATE'], errors='coerce')
```

```python
print(cleaned_df['CITYST'])
```

```python
# Check the original 'CITYST' column
print("Original 'CITYST' column:")
print(cleaned_df['CITYST'].head())

# Function to ensure the 'CITYST' column has a comma and split correctly
def split_city_state(city_state):
    if ',' in city_state:
        return city_state.split(',', 1)
    else:
        return [city_state, '']  # Return the city and an empty string for state if no comma

# Apply the function to split 'CITYST' into 'City' and 'State'
cleaned_df[['City', 'State']] = cleaned_df['CITYST'].apply(split_city_state).apply(pd.Series)

# Remove any leading/trailing whitespace from the new columns
cleaned_df['City'] = cleaned_df['City'].str.strip()
cleaned_df['State'] = cleaned_df['State'].str.strip()

# Drop the original 'CITYST' column if no longer needed
cleaned_df.drop(columns=['CITYST','City'], inplace=True)

# Verify the result
print("\nDataFrame after splitting 'CITYST' into 'City' and 'State':")
print(cleaned_df.head())
```
